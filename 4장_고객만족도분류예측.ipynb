## 피처 선택 - id 삭제, var3 은 최빈값 2로 대체

```python
cust_df.drop('ID', axis=1, inplace=True)
cust_df['var3'].replace(-999999, 2, inplace=True)

cust_df.head()
```

## X, y 데이터 분할

## 훈련 데이터 > 훈련:검증 분할

```python
X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, test_size=0.3, random_state=0)
```

## XGBoost model

```python
%%time

from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score

xgb_clf = XGBClassifier(n_estimator=500, learning_rate=0.05, random_state=156)
xgb_clf.fit(X_tr, y_tr, early_stopping_rounds=100, eval_metric='auc', eval_set=[(X_tr, y_tr), (X_val, y_val)])
xgb_roc_score = roc_auc_score(y_test, xgb_clf.predict_proba(X_test)[:, 1])
xgb_roc_score
```

## 하이퍼오피티 피라미터 평가

```python
from hyperopt.pyll.base import scope
from hyperopt import hp

search_space = {'n_estimators': scope.int(hp.quniform('n_estimators', 50, 300, 10)), 
                'max_depth': scope.int(hp.quniform('max_depth', 3, 10, 1)),
                'learning_rate': hp.loguniform('learning_rate', np.log(0.01), np.log(0.3)),
                'subsample' : hp.uniform('subsample', 0.5, 1.0),
                'colsample_bytree' : hp.uniform('colsample_bytree', 0.5, 1.0)
                }
```

```python
from hyperopt import fmin, tpe, Trials, STATUS_OK
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score

# 목적 함수 
def objective_func_xgb(params) : 

    model = XGBClassifier(
        n_estimator = 100, #params['n_estimators'],
        max_depth = params['max_depth'],
        learning_rate = params['learning_rate'],
        #subsample = params['subsample'],
        colsample_bytree = params['colsample_bytree'],
        random_state= 42,
        eval_metric = 'logloss'
    )

    score_mean = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy').mean()

    return {'loss': -1*score_mean, 'status': STATUS_OK}
```

```python
trials = Trials()

best_params = fmin(fn = objective_func_xgb,
                   space = search_space,
                   algo = tpe.suggest,
                   max_evals = 50,
                   trials = trials
                   )
```

```python
best_params
```

```python
best_model = XGBClassifier(n_estimators= int(best_params['n_estimators']), 
                           max_depth = int(best_params['max_depth']),
                           learning_rate=best_params['learning_rate'], 
                           subsample = best_params['subsample'],
                           colsample_bytree = best_params['colsample_bytree'],
                           random_state = 42,
                           eval_metric = 'logloss'
                           )
```

```python
eval_list = [(X_tr, y_tr), (X_val, y_val)]

best_model.fit(X_tr, y_tr,
               early_stopping_rounds = 100,
               eval_set = eval_list,
               eval_metric = 'auc'
               )
```

```python
from sklearn.metrics import roc_auc_score

xgb_roc_source = roc_auc_score(y_test, best_model.predict_proba(X_test)[:,1])
xgb_roc_source
```

## LGBM 하이퍼오피티 

```python
from hyperopt.pyll.base import scope
from hyperopt import hp

lgbm_search_space = {'num_leaves': hp.quniform('num_leaves', 32, 64, 1),
                     'max_depth': hp.quniform('max_depth', 100, 160, 1),
                     'min_child_samples': hp.quniform('min_child_samples', 60, 100, 1),
                     'subsample': hp.uniform('subsample', 0.7, 1),
                     'learning_rate': hp.uniform('learning_rate', 0.01, 0.2)
                     }
```

```python
from hyperopt import fmin, tpe, Trials, STATUS_OK
from lightgbm import LGBMClassifier
from sklearn.model_selection import cross_val_score

# 목적 함수 
def objective_func_lgbm(params) : 

    lgbm_clf = LGBMClassifier(
        n_estimator = 100, #params['n_estimators'],
        num_leaves= int(params['num_leaves']),
        max_depth = int(params['max_depth']),
        learning_rate = params['learning_rate'],
        subsample = params['subsample'],
        min_child_samples= int(params['min_child_samples'])
    )

    score_mean_lgbm = cross_val_score(lgbm_clf, X_train, y_train, cv=5, scoring='accuracy').mean()

    return {'loss': -1*score_mean_lgbm, 'status': STATUS_OK}
```

```python
trials_lgbm = Trials()

best_params_lgbm = fmin(fn = objective_func_lgbm,
                   space = lgbm_search_space,
                   algo = tpe.suggest,
                   max_evals = 50,
                   trials = trials_lgbm
                   )
```

```python
best_params_lgbm
```

```python
best_model_lgbm = LGBMClassifier(
        n_estimator = 100, #params['n_estimators'],
        num_leaves= int(best_params_lgbm['num_leaves']),
        max_depth = int(best_params_lgbm['max_depth']),
        learning_rate = best_params_lgbm['learning_rate'],
        subsample = best_params_lgbm['subsample'],
        min_child_samples= int(best_params_lgbm['min_child_samples'])
    )
```

```python
eval_list_lgbm = [(X_tr, y_tr), (X_val, y_val)]

best_model_lgbm.fit(X_tr, y_tr,
               early_stopping_rounds = 100,
               eval_set = eval_list_lgbm,
               eval_metric = 'auc'
               )
```

```python
from sklearn.metrics import roc_auc_score

lgbm_roc_source = roc_auc_score(y_test, best_model_lgbm.predict_proba(X_test)[:,1])
lgbm_roc_source
```
